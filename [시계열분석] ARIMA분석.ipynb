### 시계열 분해

```python
# additive, multiplicative, freq = 분기 :4, 월별 : 12, 주별 : 7, 연별 :365
result = seasonal_decompose(ts['MeanTemp'], model='additive',freq=7)
fig = plt.figure()
fig = result.plot()
fig.set_size_inches(10,8)
plt.show()
```

데이터에 패턴이 보이기 때문에 정상성이 의심된다. 따라서 이를 판단하기 위해 ACF 그래프를 그려보았다.

### 정상성 검증

```python
fig = plt.figure(figsize=(18,6))
fig = plot_acf(ts, lags=20)
plt.show()
```

- ACF 값이 아주아주 천천히 값이 작아지며 0으로 수렴하지도 않기 때문에 정상성을 만족하지 않는다는 것을 알 수 있다.   
- 따라서 단위근 검정인 ADF 검정(Augmented Dickey-Fuller test)으로 정상성을 확인해보겠다.   
- 귀무가설 : 정상성을 만족하지 않는다, 대립가설 : 정상성을 만족한다.


Augmented Dickey-Fuller (ADF) 검정
- p-value가 유의 수준보다 작으면 정상 시계열로 간주 가능
- adfuller 함수의 출력값의 1번째 요소가 p-value임

```python
result = adfuller(ts)
print('ADF Statistic : %f' % result[0])
print('p-value : %f' % result[1])
```

p-value가 0.05를 넘으므로, 귀무가설을 기각하지 못한다. 즉, 해당 데이터는 정상성을 만족하지 못한다.

### 차분 : 정상 시계열로 변환
- 평균이 일정하지 않으면 차분(difference) --> 차분해도 정상 시계열이 안되면 차분을 반복
- 분산이 일정하지 않으면 변환(transformation)

```python
ts_diff1 = ts- ts.shift() # ts.diff()
plt.plot(ts_diff1)
plt.title('ts_diff1')
plt.show()
```

```python
result = adfuller(ts_diff1[1:])
print('ADF Statistic : %f' % result[0])
print('p-value : %f' % result[1])
```

p-value가 0.05보다 작으므로 1차 차분한 데이터는 정상성을 만족한다.

### ARIMA 모델 학습 및 평가
- model fitting시 typ = 'levels'로 해주어야 한다.
- 차분이 들어간 모델의 경우 typ를 default 파라미터인 'linear'로 설정해 줄 경우 차분한 값에 대한 결과가 나온다.

```python
# fit_model
model = ARIMA(ts, order=(2,1,2))
model_fit = model.fit(disp=0)

# predict
start_index = dt.datetime(1944,6,25)
end_index = dt.datetime(1945,5,31)
forecast = model_fit.predict(start=start_index, end=end_index, typ='levels')
```

### 모델 성능 평가

```python
def scoring(y_test, y_pred):
    r2 = round(r2_score(y_test,y_pred)*100,3)
    corr = round(np.corrcoef(y_test,y_pred)[0,1],3)
    rmse = round(np.sqrt(mean_squared_error(y_test,y_pred)),3)
    
    return [r2,corr,rmse]
```

```python
print('r2, corr, rmse')
scoring(np.array(weather_bin['MeanTemp'][weather_bin['Date']>=start_index]),np.array(forecast))
```
