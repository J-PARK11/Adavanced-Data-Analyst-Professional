# 선형회귀분석 (Linear Regression)

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import  matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")
sns.set_theme(style="whitegrid")

import warnings
warnings.filterwarnings("ignore")

from sklearn.datasets import load_iris,load_breast_cancer, load_boston
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder, PolynomialFeatures
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score, r2_score
from sklearn.metrics import precision_recall_curve, roc_auc_score, roc_curve, mean_absolute_error, mean_squared_error
from sklearn.pipeline import Pipeline
```

```python
# 여러 분류 지표 출력
def get_clf_eval(y_test,pred):
    cf = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred)
    rec = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    
    print(cf)
    print('정확도 :{0:.4f}, 정밀도 :{1:.4f}, 재현율 :{2:.4f}, F1 :{3:.4f}'.format(acc,pre,rec,f1))
#     return [acc, pre, rec, f1]

# 다중 분류 지표 출력
def get_multiclf_eval(y_test,pred):
    cf = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred, average='micro')
    rec = recall_score(y_test,pred, average='macro')
    f1 = f1_score(y_test,pred, average='weighted')
    
    print(cf)
    print('정확도 :{0:.4f}, 정밀도 :{1:.4f}, 재현율 :{2:.4f}, F1 :{3:.4f}'.format(acc,pre,rec,f1))
#   return [acc, pre, rec, f1]

# roc curve
def roc_curve_plot(y_test, pred_proba):
    fprs, tprs, thresholds = roc_curve(y_test, pred_proba)
    
    plt.figure(figsize=(8,6))
    plt.plot(fprs, tprs, label='ROC')
    plt.plot([0,1],[0,1],linestyle='--', color='k', label='Random')
    
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start,end,0.1),2))
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('FPR(1 - Sensitivity)')
    plt.ylabel('TPR(Recall)')
    plt.legend()
    plt.show()
```

## 다중선형회귀 (Linear Regression)

### 데이터 불러오기, EDA

```python
# 데이터 불러오기
boston = load_boston()
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df['PRICE'] = boston.target
print("boston_df.shape :",boston_df.shape)
boston_df.head()
```

```python
# 상관관계 확인하기
fig, ax = plt.subplots(figsize=(12,8), ncols=4, nrows=2)
lm_features = ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']
for i,feature in enumerate(lm_features):
    row = int(i/4)
    col = i%4
    sns.regplot(x=feature, y='PRICE', data=boston_df, ax=ax[row][col])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(boston_df.corr(), annot=True, fmt='.2f')
plt.show()
```

```python
y_target = boston_df['PRICE']
x_data = boston_df.drop('PRICE',axis=1)
X_train, X_test, y_train, y_test = train_test_split(x_data, y_target, test_size=0.3, random_state=156)
```

### 모델 학습

```python
lr = LinearRegression(normalize=True)
lr.fit(X_train,y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
imp_series = pd.Series(index = boston.feature_names, data = lr.coef_)
top5_imp = np.abs(imp_series).sort_values(ascending=False).head()
top5_imp
```

```python
plt.title('feature importances')
sns.barplot(x = top5_imp.values, y = top5_imp.index)
plt.show()
```

```python
y_pred = lr.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('mae :',mae)
print('mse :',mse)
print('rmse :',np.sqrt(mse))
print('r2 :',r2)
```

### 교차검증

```python
lr = LinearRegression(normalize=True)
neg_mse_scores = cross_val_score(lr, X_train, y_train, cv=3, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-1*neg_mse_scores)
avg_rmse = rmse_scores.mean()

print('neg_mse_scores :',neg_mse_scores)
print('rmse_scores :',rmse_scores)
print('avg_rmse :',avg_rmse)
```

## 다항 회귀(Polynomial Regression)

```python
def polynomial_func(X):
    y = 1 + 2*X[:,0] + 3*X[:,0]**2 + 4*X[:,1]**3
    return y
```

```python
X= np.arange(4).reshape(2,2)
y = polynomial_func(X)

model = Pipeline([
    ('poly',PolynomialFeatures(degree=3)),
    ('linear',LinearRegression(normalize=True))
])

model.fit(X,y)
print('Polynomial 회귀 계수 :\n',np.round(model.named_steps['linear'].coef_,2))
```

## 규제 선형 모델

### 릿지 회귀(l2)

```python
ridge = Ridge()
neg_mse_scores = cross_val_score(ridge, X_train, y_train, scoring='neg_mean_squared_error', cv=3)
rmse_scores = np.sqrt(-1*neg_mse_scores)
avg_rmse = rmse_scores.mean()

print('neg_mse_scores :',neg_mse_scores)
print('rmse_scores :',rmse_scores)
print('avg_rmse :',avg_rmse)
```

```python
ridge = Ridge(normalize=True)
param_grid = {'alpha':[0, 0.1, 1, 10, 100]}
grid_cv = GridSearchCV(ridge, cv=3, refit=True, param_grid =param_grid, scoring='neg_mean_squared_error')
grid_cv.fit(X_train, y_train)
```

```python
print('grid_cv.best_params_ :',grid_cv.best_params_)
print('grid_cv.best_score_ :',grid_cv.best_score_)
```

```python
best_est = grid_cv.best_estimator_
print('best_est.coef_ :',np.round(best_est.coef_,2))
print('best_est.intercept_ :',best_est.intercept_)
```

```python
imp_series = pd.Series(index = boston.feature_names, data = best_est.coef_)
top5_imp = np.abs(imp_series).sort_values(ascending=False).head()
top5_imp
```

```python
plt.title('feature importances')
sns.barplot(x = top5_imp.values, y = top5_imp.index)
plt.show()
```

```python
best_pred = best_est.predict(X_test)
mse = mean_squared_error(y_test,best_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, best_pred)
r2 = r2_score(y_test, best_pred)

print('mse :',mse)
print('rmse :',rmse)
print('mae :',mae)
print('r2 :',r2)
```
