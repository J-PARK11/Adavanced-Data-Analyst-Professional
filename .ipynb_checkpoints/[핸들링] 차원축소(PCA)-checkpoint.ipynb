## 차원축소(PCA)

### ML Guide : Iris data PCA

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib as mpl
import  matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# load iris
iris = load_iris()
columns = ['sepal length', 'sepal width', 'petal length', 'petal width']
iris_df = pd.DataFrame(iris['data'], columns=columns)

print(iris_df.shape)

iris_df['target'] = iris.target
iris_df.head()
```

```python
# 상관관계 분석
colormap=plt.cm.PuBu
plt.figure(figsize=(7,7))
plt.title('Iris correlation Heatmap', size = 15, y=1.05)
sns.heatmap(iris_df.corr(), linewidths=0.1, vmax=1.0,
           square=True, cmap = colormap, linecolor='white', annot=True,
           annot_kws = {'size':16},fmt='.2g')
plt.show()
```

```python
# EDA before PCA (sepal length, sepal width)
markers=['^','s','o']

for i,marker in enumerate(markers):
    x_data = iris_df[iris_df['target']==i]['sepal length']
    y_data = iris_df[iris_df['target']==i]['sepal width']
    plt.scatter(x_data, y_data, marker = marker, label=iris.target_names[i])

plt.legend()
plt.axhline(3,linestyle = '--', color='r')
plt.axvline(6,linestyle = '--', color='r')
plt.text(6.1,4,'setosa guideline')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

Iris 데이터의 두 특성 sepal length, sepal width에 따른 데이터 분포를 살펴 본 결과, setosa는 비교적 두 특성에 대해 한 곳에 일정하게 모여있는 것을 알 수 있으나, 이외 두 품종의 경우 위 조건만으로는 분류가 힘들어보인다.

```python
# Standardization
iris_scaled = StandardScaler().fit_transform(iris.data)
print(iris_scaled.shape)
iris_scaled[:4]
```

```python
# 적절한 차원 수 튜닝
pca = PCA()
pca.fit(iris_scaled)
pca_prop = pca.explained_variance_ratio_
cumsum = np.cumsum(pca_prop)
d = np.argmax(cumsum >= 0.95) + 1

print('주성분 기여율 :',list(map(lambda x : np.round(x,2),pca_prop)))
print('누적 기여율 :',list(map(lambda x : np.round(x,2),cumsum)))
print(f'적절한 차원의 수는 : {d}')
```

```python
# 스크린 산점도
plt.plot(np.arange(len(pca_prop))+1, pca_prop,'r--')
plt.xlabel('n_comp')
plt.ylabel('pca_prop')
plt.show()
```

```python
# 누적 기여율 그래프
plt.plot(np.arange(len(cumsum))+1, cumsum,'r--')
plt.xlabel('n_comp')
plt.ylabel('cumsum')
plt.show()
```

주성분 기여율과 누적 기여율의 스크린 그래프를 분석한 결과 최적의 주성분 수는 2개인 것으로 나타났다.

```python
# PCA
pca = PCA(n_components=2)
iris_pca = pca.fit_transform(iris_scaled)

# 데이터프레임화
comp = ['comp1','comp2']
iris_df_pca = pd.DataFrame(iris_pca,columns=comp)
iris_df_pca['target'] = iris.target
print(iris_df_pca.shape)
iris_df_pca.head()
```

```python
# EDA After PCA (comp1, comp2)
markers=['^','s','o']

for i,marker in enumerate(markers):
    x_data = iris_df_pca[iris_df_pca['target']==i]['comp1']
    y_data = iris_df_pca[iris_df_pca['target']==i]['comp2']
    plt.scatter(x_data, y_data, marker = marker, label=iris.target_names[i])

plt.legend()
plt.xlabel('comp1')
plt.ylabel('sepal comp2')
plt.show()
```

```python
# PCA component Analysis
print(pca.explained_variance_ratio_)
print(list(map(lambda x : np.round(x,2),pca.explained_variance_ratio_)))
print('주성분이 설명하는 표준편차:',pca.explained_variance_)
```
