# 회귀분석(Regression Analysis)

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import  matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

import warnings
warnings.filterwarnings("ignore")

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.outliers_influence import OLSInfluence
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

## 단순선형회귀분석

```python
cars93 = pd.read_csv('data/Cars93.csv', index_col = 'Unnamed: 0')
print('cars93.shape :',cars93.shape)
cars93.head()
```

### 변수 간 상관성 검증

```python
subset = cars93[['EngineSize','Price']]
print('subset.shape :',subset.shape)
subset.head()
```

```python
# 변수 간 상관성 검증
subset.corr()
# 독립변수인 'EngineSize'와 종속변수인 'Price'간의 약한 상관성이 보임.
```

```python
colormap=plt.cm.PuBu
plt.figure(figsize=(8,8))
plt.title('correlation Heatmap', size = 15, y=1.05)
sns.heatmap(subset.corr(), linewidths=0.1, vmax=1.0,
           square=True, cmap = colormap, linecolor='white', annot=True,
           annot_kws = {'size':16}, fmt='.3f')
plt.show()
```

### 단순선형회귀분석 

```python
lm = smf.ols('Price ~ EngineSize', data = subset)
# categorical regression analysis
# model_cat = sm.OLS.from_formula('value ~ cat', df)

result = lm.fit()
print(result.summary())
```

```python
# 오차 계산
subset['pred'] = result.predict(subset['EngineSize'])
subset['error'] = np.abs(subset['Price'] - subset['pred'])
subset.head()
```

### 회귀결과 분석
- 회귀계수에 대한 t-통계량을 보면, 계수와 절편 값의 p-value 모두 유의수준 0.05보다 작으므로 도출된 회귀계수가 통계적으로 유의미하다고 볼 수 있다.   
- 요약결과 중 결정계수 부분을 보면, 결정계수가 0.357로 본 회귀모형이 전체 데이터의 약 36%을 설명할 수 있음을 알 수 있다.   
- 도출된 회귀식의 F-검정 통계량을 보면, 유의확률이 유의수준 0.05보다 훨씬 작은 것을 알 수 있다. 따라서 도출된 회귀식은 통계적으로 유의하다고 할 수 있다.   

- 회귀식 : 4.6692 + 5.5629*EngineSize = Price

### 회귀진단 (오차의 등분산성, 독립성, 정규성)

```python
# 회귀식 시각화
sns.regplot(x='EngineSize', y='Price', data=subset)
plt.text(x=2, y=50, s='y = {0}x + {1}'.format(5.662,4.669))
plt.show()
```

#### 오차의 등분산성

```python
# 잔차 등분산성
res = result.resid
plt.title('잔차 등분산성')
sns.regplot(np.arange(len(res)), res.values)
plt.show()
```

```python
plt.title('오차 등분산성')
sns.regplot(np.arange(len(subset['error'])), subset['error'])
plt.show()
```

#### 오차의 정규성

```python
# Normal Q-Q plot
sr = stats.zscore(subset['error'])
(x, y), _ = stats.probplot(sr)
plt.title('Normal Q-Q plot')
sns.regplot(x,y)
plt.plot([-3, 3], [-3, 3], '--', color='grey')
plt.show()
```

Normal Q-Q plot은 표준화된 잔차의 확률도이다. 정규성 가정을 만족한다면 그래프의 점들은 45도 각도의 직선을 이루어야 한다.

```python
# 정규성 테스트
k2, p = stats.normaltest(subset['error'])
print(p)
```

```python
# 샤피로-윌크 테스트
k2, p = stats.shapiro(subset['error'])
print(p)
```

정규성 테스트, shapiro 테스트 검정 결과, 유의확률이 유의수준 0.05보다 작으므로 오차의 정규성이 보장된다.

#### 오차의 독립성

```python
# 더빈-왓슨 검정 : 
durbin_watson(subset['error'])
# 더빈왓슨 검정통계량이 0.8로 0에 가까움으로 양의 상관을 나타낸다.
```

#### 극단값 및 관측치의 영향도 검정

```python
# Cooks's Distance
cd, _ = OLSInfluence(result).cooks_distance
cd.sort_values(ascending=False).head()
```

### 회귀모형을 활용한 예측

```python
lr_pred = result.predict(subset['EngineSize'].head())
df_pred = pd.concat([subset['Price'].head(),lr_pred],axis=1)
df_pred.rename(columns = {0:'pred'}, inplace=True)
df_pred['error'] = np.abs(df_pred['Price']-df_pred['pred'])
df_pred
```

## 다중선형회귀분석

## 1. 범주형 변수 처리를 활용한 Iris 회귀분석

```python
# Iris data load
iris = load_iris()
iris_df=pd.DataFrame(iris.data, columns=[word[:12].strip().replace(' ','_') for word in iris.feature_names])
iris_df['Species'] = iris.target_names[iris.target]
print('iris_df.shape :',iris_df.shape)
iris_df.head()
```

```python
# 더비변수 정의 (dummy variable)
iris_new = pd.concat([iris_df,pd.get_dummies(iris_df['Species'])], axis=1)
iris_new.drop('Species',axis=1,inplace=True)
print('iris_new.shape :',iris_new.shape)
iris_new.head()
```

### 변수 간 상관성 및 다중공선성 검정

```python
iris_new.corr()
```

```python
colormap=plt.cm.PuBu
plt.figure(figsize=(12,12))
plt.title('correlation Heatmap', size = 15, y=1.05)
sns.heatmap(iris_new.corr(), linewidths=0.1, vmax=1.0,
           square=True, cmap = colormap, linecolor='white', annot=True,
           annot_kws = {'size':16}, fmt='.3f')
plt.show()
```

우선, petal_length 종속변수와 다른 독립변수 간의 강한 선형성이 존재함을 알 수 있다.

```python
# 다중선형회귀분석
lm = smf.ols('petal_length ~' + '+'.join(iris_new.columns.difference(['petal_length'])), data = iris_new)
# lm = sm.OLS(Y,X)

result = lm.fit()
print(result.summary())
```

```python
# 오차분석
subset = iris_new.drop('petal_length',axis=1)
subset['pred'] = result.predict(subset)
subset['error'] = np.abs(subset['pred']-iris_new['petal_length'])
subset.head()
```

### 회귀결과 분석
- 회귀계수에 대한 t-통계량을 보면, 모든 계수와 절편 값의 p-value 모두 유의수준 0.05보다 작으므로 도출된 회귀계수가 통계적으로 유의미하다고 볼 수 있다.   
- 요약결과 중 결정계수 부분을 보면, 결정계수가 0.979로 본 회귀모형이 전체 데이터의 약 98%을 설명할 수 있음을 알 수 있다.   
- 도출된 회귀식의 F-검정 통계량을 보면, 유의확률이 유의수준 0.05보다 훨씬 작은 것을 알 수 있다. 따라서 도출된 회귀식은 통계적으로 유의하다고 할 수 있다.

### 회귀진단 (다중공선성, 오차의 등분산성|정규성|독립성)

#### 다중공선성 검사 : -> 회귀계수가 무의미할 떄
사실 회귀계수가 다 유의미하다고 나오면 공선성이 있어도 제거할 필요는 없다. 하지만, 만약 회귀계수가 무의미한 특성의 경우 VIF가 10이상이면 제거 그리고 그렇지 않더라도 제거해도 된다.

```python
print('독립변수 리스트 :',lm.exog_names)
```

```python
# 분산팽창요인(VIF) 출력 함수 정의
def VIF(model):
    for i,col in enumerate(model.exog_names):
        r = variance_inflation_factor(model.exog, i)
        print(col,r)
        
VIF(lm)
# petal_width의 VIF가 18으로 10보다 높아 다중공선성이 매우 크지만, 회귀계수가 유의미하므로 따로 처리해주지는 않았다.
```

#### 오차의 등분산성

```python
# 잔차 등분산성
res = result.resid
plt.title('잔차 등분산성')
sns.regplot(np.arange(len(res)), res.values)
plt.show()
```

```python
plt.title('오차 등분산성')
sns.regplot(np.arange(len(subset['error'])),subset['error'])
plt.show()
```

#### 오차의 정규성

```python
# Normal Q-Q plot
sr = stats.zscore(subset['error'])
(x, y), _ = stats.probplot(sr)
plt.title('Normal Q-Q plot')
sns.regplot(x,y)
plt.plot([-3, 3], [-3, 3], '--', color='grey')
plt.show()
```

```python
# 정규성 테스트
k2, p = stats.normaltest(subset['error'])
print(p)
```

```python
# 샤피로-윌크 테스트
k2, p = stats.shapiro(subset['error'])
print(p)
```

#### 오차의 독립성

```python
# 더빈-왓슨 검정 : 
durbin_watson(subset['error'])
# 더빈왓슨 검정통계량이 0.8로 0에 가까움으로 양의 상관을 나타낸다.
```

#### 극단값 검정 관측치의 영향도 검정

```python
# Cooks's Distance
cd, _ = OLSInfluence(result).cooks_distance
cd.sort_values(ascending=False).head()
```

### 회귀모형을 활용한 예측

```python
aa = pd.concat([iris_new,subset[['pred','error']]],axis=1)
aa[['petal_length','pred','error']]
```
