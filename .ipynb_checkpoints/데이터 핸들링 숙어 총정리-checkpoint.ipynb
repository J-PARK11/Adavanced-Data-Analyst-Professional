## 주택 가격 데이터를 바탕으로 한 데이터 전처리
OneHotEncoding, SelectKBest, train_test_split 

```python
housing_df = pd.read_csv('data/house_price.csv')
housing_df.dropna(inplace=True)
housing_df.drop(columns=['Date','Suburb','Address','SellerG'],inplace=True)
print(housing_df.shape)
housing_df.head()
```

```python
housing_df.info()
```

OneHotEncoding

```python
df = housing_df[housing_df.columns.difference(['Type','Method','Regionname','CouncilArea'])]
dummy_col = housing_df[['Type','Method','Regionname','CouncilArea']]
dummy_col = pd.get_dummies(dummy_col)
df = pd.concat([df,dummy_col],axis=1)
print(df.shape)
df.head()
```

Train_Test_Split

```python
X = df[df.columns.difference(['Price'])]
y = df['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2)
print(X_train.shape)
print(X_test.shape)
```

SelectKBest   
scikit-learn의 SelectKBest 모듈은 target 변수와 그외 변수 사이의 상관관계를 계산하여 가장 상관관계가 높은 변수 k개를 선정할 수 있는 모듈입니다. 상관관계를 분석하는 방법은 f-regression 방식과 chi2(카이제곱)방식, f-classif 등이 있습니다.

```python
selector = SelectKBest(score_func = f_regression, k=20)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)
print('X_train_selected:',X_train_selected.shape)
print('X_test_selected:',X_test_selected.shape)
```

```python
# SelectKBest 모듈 상세정보
all_names = X_train.columns
selected_mask = selector.get_support()
selected_names = all_names[selected_mask]
print(selected_names)
```

### 이상치 처리

```python
def outlier_iqr(data, column):
    
    q25, q75 = np.quantile(data[column], 0.25), np.quantile(data[column], 0.75)
    iqr = q75 - q25
    cut_off = iqr*1.5
    lower_line, upper_line = q25 - cut_off, q75 + cut_off
    
    return lower_line, upper_line
```

```python
lower_line, upper_line = outlier_iqr(housing_df,'Price')
print('upper_line:',upper_line)
print('lower_line:',lower_line)
```

```python
sns.boxplot(housing_df['Price'])
plt.title('housing_df Price boxplot',y=1.05, size=15)
plt.axvline(x=upper_line, linestyle='--', color='red')
plt.show()
```

Trim or Winsorizing(Cliping)

```python
trim_price = housing_df['Price'].copy()
upper_index = trim_price[trim_price > upper_line].index
lower_index = trim_price[trim_price < lower_line].index

trim_price = trim_price.drop(index = upper_index)
trim_price = trim_price.drop(index = lower_index)

plt.title('Trimming', y=1.05, size=15)
sns.boxplot(trim_price)
plt.show()
```

```python
clip_price = housing_df['Price'].copy()
upper_index = clip_price[clip_price > upper_line].index
lower_index = clip_price[clip_price < lower_line].index

clip_price[upper_index] = upper_line
clip_price[lower_index] = lower_line

plt.title('Cliping', y=1.05, size=15)
sns.boxplot(clip_price)
plt.show()
```

```python
clip_price = housing_df['Price'].copy()

clip_price = clip_price.clip(lower_line, upper_line)

plt.title('Cliping', y=1.05, size=15)
sns.boxplot(clip_price)
plt.show()
```
