# Logistic Regression (로지스틱 회귀분석)

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import  matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")
sns.set_theme(style="whitegrid")

import warnings
warnings.filterwarnings("ignore")

from sklearn.datasets import load_iris,load_breast_cancer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score
from sklearn.metrics import precision_recall_curve, roc_auc_score, roc_curve
```

```python
# 여러 분류 지표 출력
def get_clf_eval(y_test,pred):
    cf = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred)
    rec = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    
    print(cf)
    print('정확도 :{0:.4f}, 정밀도 :{1:.4f}, 재현율 :{2:.4f}, F1 :{3:.4f}'.format(acc,pre,rec,f1))
#     return [acc, pre, rec, f1]

# 다중 분류 지표 출력
def get_multiclf_eval(y_test,pred):
    cf = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred, average='micro')
    rec = recall_score(y_test,pred, average='macro')
    f1 = f1_score(y_test,pred, average='weighted')
    
    print(cf)
    print('정확도 :{0:.4f}, 정밀도 :{1:.4f}, 재현율 :{2:.4f}, F1 :{3:.4f}'.format(acc,pre,rec,f1))
#   return [acc, pre, rec, f1]

# roc curve
def roc_curve_plot(y_test, pred_proba):
    fprs, tprs, thresholds = roc_curve(y_test, pred_proba)
    
    plt.figure(figsize=(8,6))
    plt.plot(fprs, tprs, label='ROC')
    plt.plot([0,1],[0,1],linestyle='--', color='k', label='Random')
    
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start,end,0.1),2))
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('FPR(1 - Sensitivity)')
    plt.ylabel('TPR(Recall)')
    plt.legend()
    plt.show()
```

## 이항 로지스틱 회귀분석

```python
# 데이터 불러오기
cancer = load_breast_cancer()
cancer_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
print('cancer_df.shape :',cancer_df.shape)
cancer_df.head()
```

### 데이터 전처리

```python
# 표준화
scaler = StandardScaler()
cancer_scaled = scaler.fit_transform(cancer_df)
```

```python
# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(cancer_scaled, cancer.target, test_size=0.3,
                                                    random_state=0, shuffle=True, stratify=cancer.target)
print("X_train.shape :",X_train.shape)
print("X_test.shape :",X_test.shape)
```

### 모델 학습하기

```python
# 모델 학습
lr_clf = LogisticRegression(class_weight='balanced', solver='lbfgs', multi_class='auto')
lr_clf.fit(X_train,y_train)
```

```python
# 특성별 회귀 계수
print(lr_clf.coef_)
print(lr_clf.intercept_)
```

```python
# 변수 중요도
feature_coef = pd.Series(np.abs(lr_clf.coef_[0]),cancer.feature_names).sort_values(ascending=False)
feature_coef[:5]
```

```python
# 회귀 계수 시각화
plt.figure(figsize=(20,8))
plt.title('Absolute regression coef',size=20)
sns.barplot(feature_coef.index,feature_coef.values)
plt.xticks(rotation=30)
plt.show()
```

### 테스트 데이터 평가

```python
# 테스트 데이터 평가.
pred = lr_clf.predict(X_test)
get_clf_eval(y_test,pred)
```

```python
# roc curve & roc_auc score
print('roc_auc_score :',roc_auc_score(y_test,pred))
roc_curve_plot(y_test, pred)
```

### 하이퍼파라미터튜닝

```python
# GridSearchCV
model = LogisticRegression(class_weight = 'balanced', multi_class = 'auto')
param_grid = {'penalty':['l1','l2','elasticnet'],
              'C':[0.01, 0.1, 1, 5, 10],
              'solver':['saga','lbfgs']}

lr_grid = GridSearchCV(model, cv=3, refit=True, param_grid = param_grid)
lr_grid.fit(X_train,y_train)
```

```python
score_df = pd.DataFrame(lr_grid.cv_results_)
score_df[['params','mean_test_score','rank_test_score','split0_test_score']].sort_values('rank_test_score').head()
```

```python
# best parameter & score
print(lr_grid.best_params_)
print(lr_grid.best_score_)
```

```python
# best estimator prediction
best = lr_grid.best_estimator_
best_pred = best.predict(X_test)
```

```python
# best estimator performance
get_clf_eval(y_test,best_pred)
```

```python
# roc curve & roc_auc score
pred_proba = best.predict_proba(X_test)[:,1]
print('roc_auc_score :',roc_auc_score(y_test,pred_proba))
roc_curve_plot(y_test, pred_proba)
```

### 다항 로지스틱 회귀분석

```python
# 데이터 불러오기
iris= load_iris()
iris_df = pd.DataFrame(iris.data, columns=[word[:12].strip() for word in iris.feature_names])
print('iris_df.shape :',iris_df.shape)
iris_df.head()
```

### 데이터 전처리

```python
# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris.data)
```

```python
# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_scaled, iris.target, random_state=42, shuffle=True,
                                                   test_size=0.2, stratify=iris.target)
```

### 모델 학습

```python
# 다중 로지스틱 회귀 수행
lr_clf = LogisticRegression(class_weight = 'balanced', multi_class='multinomial')
lr_clf.fit(X_train,y_train)
```

```python
# 특성별 회귀 계수
print(lr_clf.coef_)
print(lr_clf.intercept_)
```

```python
# 변수 중요도
feature_coef = pd.Series(np.abs(lr_clf.coef_[0]),iris.feature_names).sort_values(ascending=False)
feature_coef[:5]
```

```python
# 회귀 계수 시각화
plt.figure(figsize=(8,6))
plt.title('Absolute regression coef',size=20)
sns.barplot(feature_coef.index,feature_coef.values)
plt.show()
```

### 모델 평가

```python
pred = lr_clf.predict(X_test)
get_multiclf_eval(y_test,pred)
```

### 하이퍼파라미터튜닝

```python
# GridSearchCV
model = LogisticRegression(class_weight = 'balanced', multi_class = 'multinomial')
param_grid = {'penalty':['l1','l2','elasticnet'],
              'C':[0.01, 0.1, 1, 5, 10],
              'solver':['saga','lbfgs']}

lr_grid = GridSearchCV(model, cv=3, refit=True, param_grid = param_grid)
lr_grid.fit(X_train,y_train)
```

```python
score_df = pd.DataFrame(lr_grid.cv_results_)
score_df[['params','mean_test_score','rank_test_score','split0_test_score']].sort_values('rank_test_score').head()
```

```python
print(lr_grid.best_params_)
print(lr_grid.best_score_)
```

```python
best = lr_grid.best_estimator_
best_pred = best.predict(X_test)
```

```python
get_multiclf_eval(y_test,best_pred)
```
