# 부스팅(GBM, XGB, LGBM)

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import  matplotlib.pyplot as plt
import seaborn as sns

import time
import datetime as dt

mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")
sns.set_theme(style="whitegrid")

import warnings
warnings.filterwarnings("ignore")

from sklearn.datasets import load_iris,load_breast_cancer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score
from sklearn.metrics import precision_recall_curve, roc_auc_score, roc_curve, plot_roc_curve

from xgboost import XGBClassifier, plot_importance
from lightgbm import LGBMClassifier, plot_importance
```

```python
# 여러 분류 지표 출력
def get_clf_eval(y_test,pred):
    cf = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred)
    rec = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    
    print(cf)
    print('정확도 :{0:.4f}, 정밀도 :{1:.4f}, 재현율 :{2:.4f}, F1 :{3:.4f}'.format(acc,pre,rec,f1))
#   return [acc, pre, rec, f1]

# 다중 분류 지표 출력
def get_multiclf_eval(y_test,pred):
    cf = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred, average='micro')
    rec = recall_score(y_test,pred, average='macro')
    f1 = f1_score(y_test,pred, average='weighted')
    
    print(cf)
    print('정확도 :{0:.4f}, 정밀도 :{1:.4f}, 재현율 :{2:.4f}, F1 :{3:.4f}'.format(acc,pre,rec,f1))
#   return [acc, pre, rec, f1]

# roc curve
def roc_curve_plot(y_test, pred_proba):
    fprs, tprs, thresholds = roc_curve(y_test, pred_proba)
    
    plt.figure(figsize=(8,6))
    plt.plot(fprs, tprs, label='ROC')
    plt.plot([0,1],[0,1],linestyle='--', color='k', label='Random')
    
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start,end,0.1),2))
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('FPR(1 - Sensitivity)')
    plt.ylabel('TPR(Recall)')
    plt.legend()
    plt.show()
```

## GBM(Gradient Boosting Machine)

```python
# 데이터 불러오기
cancer = load_breast_cancer()
cancer_df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
print('cancer_df.shape :',cancer_df.shape)
cancer_df.head()
```

```python
# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2, shuffle=True,
                                                   random_state=42, stratify=cancer.target)
```

```python
# GBM 하이퍼파라미터 튜닝

param_grid = {
    'learning_rate':[0.05,0.1],
    'n_estimators':[50,100]
}

gbm_clf = GradientBoostingClassifier(max_depth = 6, min_samples_leaf = 8, min_samples_split = 8)
grid_cv = GridSearchCV(gbm_clf, cv=2, refit=True, param_grid = param_grid)
grid_cv.fit(X_train, y_train)
```

```python
print('grid_cv.best_params_ :',grid_cv.best_params_)
print("grid_cv.best_score_ :",grid_cv.best_score_)
```

```python
best = grid_cv.best_estimator_
pred = best.predict(X_test)
get_clf_eval(y_test,pred)
```

## XGB(eXtra Gradient Boosting)

```python
# xgb 분류기 학습
XGB_clf = XGBClassifier(n_estimators=400, learning_rate=0.1, max_depth = 3, subsample=0.8)
evals = [(X_test,y_test)]

XGB_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='logloss', eval_set=evals, verbose=True)
```

```python
XGB_clf.feature_importances_
```

```python
pred = XGB_clf.predict(X_test)
get_clf_eval(y_test,pred)
```

```python
# 하이퍼파라미터 튜닝
model = XGBClassifier(n_estimators=400, learning_rate=0.1, max_depth = 3)

param_grid = {
    'learning_rate':[0.01,0.05,0.1],
    'n_estimators':[100,200,300,400],
}

grid_cv = GridSearchCV(model, cv=3, refit=True, param_grid=param_grid)
grid_cv.fit(X_train, y_train, early_stopping_rounds=50, eval_metric='logloss', eval_set=evals, verbose=True)
```

```python
print('grid_cv.best_params_ :',grid_cv.best_params_)
print("grid_cv.best_score_ :",grid_cv.best_score_)
```

```python
best = grid_cv.best_estimator_
pred = best.predict(X_test)
get_clf_eval(y_test,pred)
```

```python
fig, ax = plt.subplots(figsize=(10,12))
plot_importance(best, ax=ax)
plt.show()
```

## LGBM(Light Gradient Boosting Machine)

```python
lgbm_clf = LGBMClassifier(n_estimators=400, learning_rate=0.1, subsample=0.8, max_depth=15, min_child_samples=25)
evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train, early_stopping_rounds=50, eval_metric='log_loss', eval_set = evals)
```

```python
lgbm_clf.feature_importances_
```

```python
pred = XGB_clf.predict(X_test)
get_clf_eval(y_test,pred)
```

```python
fig, ax = plt.subplots(figsize=(10,12))
plot_importance(lgbm_clf, ax=ax)
plt.show()
```
