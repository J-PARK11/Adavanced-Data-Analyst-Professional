# 교차분석 & 상관분석

## 교차분석(CrossValidation)

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import  matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

import warnings
warnings.filterwarnings("ignore")

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from scipy import stats
```

### 적합성 검정

```python
survey = pd.read_csv('data/survey.csv', index_col='Unnamed: 0')
print('survey.shape :',survey.shape)
survey.head()
```

```python
# 교차표 (Crosstable)
ct = survey['W.Hnd'].value_counts().to_frame().T
ct.rename(index={'W.Hnd':''},inplace=True)
print(ct)
```

```python
# 적합성 검정
stats.chisquare(ct.iloc[0].values, f_exp=(0.8,0.2))
# 유의확률이 유의수준 0.05보다 작으므로 표본이 위 분포를 따르지 않을 것이다.
```

### 독립성(동질성) 검정

```python
# 독립성 검정 = 동질성 검정
ct2 = pd.crosstab(index = survey['W.Hnd'], columns=survey['Exer'])
ct2
```

```python
stat, p_value, _,_ = stats.chi2_contingency(ct2)
print(p_value)
# 카이제곱 독립성 검정의 유의확률이 0.54로 유의수준 0.05보다 높으므로
# 두 범주형 변수 간의 독립성이 보장된다.
```

## 상관분석(Correlation)

```python
aq = pd.read_csv('data/airquality.csv')
air = aq.iloc[:,:-2]
air.dropna(how='any', axis=0, inplace=True)
print('air.shape: ',air.shape)
air.head()
```

```python
# 상관계수 산출 : pearson
air.corr(method='pearson')
```

```python
# 상관계수 산출 : spearman
air.corr(method='spearman')
```

```python
# 상관계수 산출 : kendall
air.corr(method='kendall')
```

```python
# 상관분석
x1, x2 = air['Ozone'], air['Wind']
print(stats.pearsonr(x1, x2))
print(stats.spearmanr(x1, x2))
print(stats.kendalltau(x1, x2))
# 세 상관분석 검정 방식 결과 모두 유의확률이 유의수준 0.5보다 작으므로
# 두 변수 간의 상관관계는 존재하며, 약한 음의 상관관계를 띄는 것을 알 수 있다.
```

### 상관관계 시각화

```python
colormap=plt.cm.PuBu
plt.figure(figsize=(8,8))
plt.title('airquality correlation Heatmap', size = 15, y=1.05)
sns.heatmap(air.corr(), linewidths=0.1, vmax=1.0,
           square=True, cmap = colormap, linecolor='white', annot=True,
           annot_kws = {'size':16})
plt.show()
```

```python
ax = plt.figure(figsize=(6,6))
ax = sns.pairplot(air)
plt.show()
```

```python
g = sns.pairplot(air, diag_kind = 'kde')
g.map_lower(sns.kdeplot, levels=4, color=".2")
plt.show()
```
